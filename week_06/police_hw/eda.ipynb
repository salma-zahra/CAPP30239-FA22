```python
import pandas as pd
import numpy as np
import datetime as dt 
```

```python
df = pd.read_json('a3cleanedonly2015.json')
# df.head(20)
```

```python
len(df['UID'].unique())
```

```python
# Number of races 

print(df['Race'].unique())
df['Race'].value_counts()
```

Data cleaning: Dropped column 'Unamed: 0' and 'FIELD1'. Using UID as the unique identifier. 
We have 1537 data points. 

114 missing points for Race (7% of the data). We converted it to 'Unknown'.

```python
df.drop(columns = ['FIELD1', 'Unnamed: 0'], inplace = True)
```

```python
df['Race'] = np.where(df['Race'] == '', 'Unknown', df['Race'])
df['Race'].value_counts()
```

```python
df.head(10)
```

```python
# Grouping data by month 

# Extract month from Date 

df['Month'] = df['Date'].dt.month

# df
```

```python
# Group by month 
monthly_shootings = df.groupby(['Month', 'Race'])['UID'].count().reset_index()
monthly_shootings.rename(columns = {'UID' : 'Count'}, inplace = True)
monthly_shootings
```

```python
month_and_race = monthly_shootings.pivot(index='Month', columns='Race', values='Count')
month_and_race
```
